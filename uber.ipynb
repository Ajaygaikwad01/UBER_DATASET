```python
import pandas as pd
```

```python
df=pd.read_csv("uber.csv")
```

```python
df.head(5)
```

```python
df["MILES*"]=df["MILES*"].astype(int)
```

```python
def trip(x):
    if x > 10 and x < 30:
        return "long Trip"
    elif x <10:
        return "small Trip"
    elif x > 30:
        return " long Trip"
        

df["trip"]=df["MILES*"].apply(trip)
```
